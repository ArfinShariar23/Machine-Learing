```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('Canada4.csv')
```

```python
df
```

```python
#data visulaization
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income $')
plt.title('Canda Per year Per Capita Income')
plt.scatter(df.year,df.PCI,color = 'red',marker = '+')
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['year']],df.PCI)
```

```python
reg.predict([[2022]])
```
